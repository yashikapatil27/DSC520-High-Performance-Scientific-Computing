### OPENMP

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'code/time_omp.txt'
data = pd.read_csv(file_path, delimiter='\s+', header=None, names=['Processes','Time'])

# Calculate speedup
time_ratio = data.loc[data['Processes'] == 2, 'Time'].values[0]
data['Speedup'] = time_ratio / data['Time']

# Fit a linear regression line to the speedup data
coefficients = np.polyfit(data['Processes'].values, data['Speedup'].values, 1)
slope = coefficients[0]

plt.figure(figsize=(8, 6))
plt.plot(data['Processes'].values, data['Speedup'].values, marker='o', linestyle='-', color='b', label='Speedup')
plt.title('Speedup vs. Number of Cores')
plt.xlabel('Number of Processes')
plt.ylabel('Speedup')
plt.grid(True)
plt.legend()

# Print the slope
print(f"Slope of the linear fit: {slope:.6f}")

plt.show()
```
